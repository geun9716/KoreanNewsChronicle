````python
import os
import pandas as pd

class Reader:
    def __init__(self, filePath):
        newses=[]
        news=pd.read_csv(filePath,encoding="UTF-8")
        newses.append(news)
        self.docs=pd.concat(newses,ignore_index=True)
    
    def read(self, id):
        return str(self.docs['text'][id])

reader = Reader('../Data/2020(1_10).csv')

print(reader.docs)
````

````python
from soynlp import DoublespaceLineCorpus
from soynlp.word import WordExtractor

corpus = DoublespaceLineCorpus("../Data/datas.txt")
len(corpus)
````

````python
word_extractor = WordExtractor()
word_extractor.train(corpus)
word_score_table = word_extractor.extract()
````

````python
from soynlp.utils import DoublespaceLineCorpus
from soynlp.noun import LRNounExtractor_v2

corpus_path = "../Data/datas.txt"
sents = DoublespaceLineCorpus(corpus_path, iter_sent=False)
print(sents)
for n_doc, doc in enumerate(sents):
    continue
print('num docs = {}'.format(n_doc+1))

noun_extractor = LRNounExtractor_v2(verbose=True)
nouns = noun_extractor.train_extract(sents)
````

````python
from pprint import pprint as pp

pp(nouns.keys())
````

````python
file = open("한국어불용어100.txt", 'r', encoding="utf-8")
stopword=[]

while True:
    line = file.readline()
    if not line: break
    wordlist = line.split('\t')
    if (wordlist[1].startswith('N')):
        stopword.append(wordlist[0])

print(stopword)
````

````python
result = []

for w in nouns.keys():
    if w not in stopword:
        result.append(w)

print(result)
````

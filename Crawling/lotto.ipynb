```python
import pandas as pd

file = '2019_LDA_topic3.csv'
```

```python
headline_droplist = ['로또', '복권', '부고']

df = pd.read_csv(file, encoding="utf-8-sig")
df
```

```python
for index in range(0, len(df)):
    for drop in headline_droplist:
        if drop in df['headline'][index]:
            df['headline'][index] = None
            break

df = df.dropna(axis=0)
df = df.reset_index(drop=True)

df.to_csv("pre_"+file, encoding="utf-8-sig", mode = 'w', index=False) # 저장시 파일 이름 주의
```

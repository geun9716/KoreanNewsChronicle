```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from collections import Counter

plt.close()
months = mdates.MonthLocator()
df = pd.read_csv("2019_fix.csv", encoding="utf-8")
```

```python
labels_cnt = sorted(Counter(df['labels']).most_common(15))
print(labels_cnt)
```

```python
labels_cnt = sorted(Counter(df['labels']).most_common(15))

result = []

fig, ax = plt.subplots(figsize=(25, 5))
for labels in labels_cnt:
    tmp = df[df['labels'] == labels[0]]
    result = sorted(Counter(tmp['date']).most_common())
    s = pd.DataFrame(result, columns=['date', 'cnt'])
    s = s.sort_values('date')
    ax.plot_date(s['date'], s['cnt'], 'o', label=labels[0])

ax.set(title='hot_topic', xlabel='Date', ylabel='cnt')
ax.xaxis.set_major_locator(months)
ax.format_xdata = mdates.DateFormatter('%y-%m')
ax.grid(True)
ax.legend(loc = 1)
# fig.autofmt_xdate()
plt.xticks(rotation=45)
plt.show()
```

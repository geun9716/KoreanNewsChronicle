```python
import pandas as pd
import numpy as np
df = pd.read_csv("InTime_News.csv", encoding='utf-8-sig')
df
```

```python
count = []
df['date'] = df['date'].apply(lambda x:pd.to_datetime(x, format='%Y-%m-%d'))
grouped = list(df[['cId', 'date']].groupby('cId'))
for i in range(len(grouped)):
    tmp = grouped[i][1].set_index('date').resample('1M').sum()
    tmp1 = tmp.sort_values(by=['cId'], ascending=False).index[0]
    tmp2 = str(tmp1.year) + '-0' + str(tmp1.month)
    count.append(tmp2)

result = pd.DataFrame(count, columns=['cluster_date'])
result.to_csv("cluster_date.csv", encoding='utf-8-sig', index=False)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import csv
import pandas as pd
import sys
from collections import Counter

df = pd.read_csv("../Data/2019/2019_DBSCAN_new3.csv")
df = df.sort_values(by=['labels'], axis = 0)
```

```python
last_index = 0

for i in range(50):
    noun_count = {}
    for j in range(last_index, len(df), 1):
        if(int(df.iloc[j]['labels']) == i):
            N_list = df.iloc[j]['N_list'].replace("'", "").replace("[", "").replace("]", "").split(', ')
            for noun in N_list:
                if noun in noun_count:
                    noun_count[noun] += 1
                else:
                    noun_count[noun] = 1
        else:
            last_index = j
            break
            
    noun_count = sorted(noun_count.items(), reverse=True, key=lambda item: item[1])
    print(noun_count[:20])
    
    wordcloud = WordCloud(font_path='malgun.ttf', background_color='white').generate_from_frequencies(dict(noun_count[:20]))
    plt.figure(figsize=(30,30)) #이미지 사이즈 지정
    plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
    plt.axis('off') #x y 축 숫자 제거
    
    filename = '2019-{0:03d}.png'.format(i+1)
    plt.savefig(filename)
```

```python
import pandas as pd
import collections
df = pd.read_csv("../Data/2019/201901_DBSCAN_CV.csv")
```

```python
counts = []
for i in range(df['labels'].max()):
    mydict = {}
    labels = df[df['labels'] == i]
    word_list = []
    for topics in labels['topics']:
        topics = topics.split(", ")
        for topic in topics:
            word_list.append(topic)
        counter = collections.Counter(word_list)
    counts.append(counter)
```

```python
for count in counts:
    print(count, "\n\n\n\n")
```
